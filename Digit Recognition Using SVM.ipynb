<h1>Digit Recognition Using Support Vector Machine (SVM)</h1>

```python
#Importing Libraries
import numpy as np
from sklearn.datasets import load_digits
```

```python
#Loading the Dataset 
df = load_digits()
```

```python
#Summarize the Dataset
print("DataSet - Input X\n", df.data)
print()

print("Target - Output Y\n", df.target)
print()

print("Shape\n", df.data.shape)
print()

print("Image Shape\n", df.images.shape)
print()

dataimagelength = len(df.images)
print("Number of Images available\n", dataimagelength)
```

```python
#Visualize the Dataset

n = 3 #No. of sample out of samples total 1797

import matplotlib.pyplot as plt
plt.gray()
plt.matshow(df.images[n])
plt.show()

df.images[n] #image in a number form
```

```python
#Segregate dataset into X(Input Variable) & Y(Output Variable)

x = df.images.reshape((dataimagelength, -1)) #Using only Images Data except the classes data
print("Input Variable - X\n", x)
print()

y = df.target
print("Output Variable - Y\n", y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
print("X_Train\n", x_train.shape)
print()
print("X_Test\n", x_test.shape)
print()
print("Y_Train\n", y_train.shape)
print()
print("Y_Test\n", y_test.shape)
```

```python
#Training

from sklearn import svm
model = svm.SVC(kernel = 'linear') #From Support Vector Machine called Support Vector Classifier
model.fit(x_train, y_train)
```

```python
#Prediction, What the digit is from Test Data

n = 1
result = model.predict(df.images[n].reshape((1, -1))) #Predict result for image n using model.predict() and reshape().
plt.imshow(df.images[n], cmap=plt.cm.gray_r, interpolation='nearest') #Uses the imshow() function from plt to display the image at index n with a grayscale color map (cmap=plt.cm.gray_r) and nearest-neighbor interpolation.
print(result)
print("\n")
plt.axis('off') #Removes the axis from the plot using plt.axis('off').
plt.title('%i' %result) #Sets the title of the plot to the predicted result using plt.title('%i' %result).
plt.show()
```

```python
# Prediction for Test Data (Code perform to check the Predication and Actual Value format)

y_pred = model.predict(x_test) #The model.predict(x_test) function is called to predict the output values based on the input x_test. The predicted values are stored in the variable y_pred.
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1)) #The np.concatenate() function is used to concatenate the predicted values (y_pred) and the actual values (y_test) along the second axis (axis=1). This creates a new array where each row contains the predicted value and the corresponding actual value.
```

```python
# Evaluate Model_Accuracy Score

from sklearn.metrics import accuracy_score
print("Accuracy of the Model: {0}%".format(accuracy_score(y_test, y_pred)*100))
```

```python
from sklearn import svm
model1 = svm.SVC(kernel='linear')
model2 = svm.SVC(kernel='rbf')
model3 = svm.SVC(gamma=0.001)
model4 = svm.SVC(gamma=0.001, C=0.1)

model1.fit(x_train, y_train)
model2.fit(x_train, y_train)
model3.fit(x_train, y_train)
model4.fit(x_train, y_train)

y_predModel1 = model1.predict(x_test)
y_predModel2 = model2.predict(x_test)
y_predModel3 = model3.predict(x_test)
y_predModel4 = model4.predict(x_test)

print("Accuracy of the Model 1: {0}%".format(accuracy_score(y_test, y_predModel1)*100))
print("Accuracy of the Model 2: {0}%".format(accuracy_score(y_test, y_predModel2)*100))
print("Accuracy of the Model 3: {0}%".format(accuracy_score(y_test, y_predModel3)*100))
print("Accuracy of the Model 4: {0}%".format(accuracy_score(y_test, y_predModel4)*100))
```
